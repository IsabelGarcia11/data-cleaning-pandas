```python
rango_edad=[str(i) for i in range(0,100)]
```

```python
def Ages(x):
    if str(x) in rango_edad:
        return x
    elif str(x).__contains__("months"):
        return "3"
    elif str(x).__contains__("&"):
        return x
    elif str(x).__contains__("to"):
        return x
    elif str(x).__contains__("or"):
        return x
    else:
        return "No_Info"
```
